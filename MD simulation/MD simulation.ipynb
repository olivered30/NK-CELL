```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from pandas import DataFrame
```

```python
LLL= pd.read_csv("https://raw.githubusercontent.com/olivered30/NK-CELL/main/MD%20simulation/LLL.csv")
```

```python
VLL=pd.read_csv("https://raw.githubusercontent.com/olivered30/NK-CELL/main/MD%20simulation/VLL.csv")
```

```python
MD=pd.read_csv("https://raw.githubusercontent.com/olivered30/NK-CELL/main/MD%20simulation/Combined.csv")
```

```python
a=MD["CD94-LLL"]
b=MD["CD94-VLL"]
c=MD["NKG2A-LLL"]
d=MD["NKG2A-VLL"]
e=MD["CD94/NKG2A-LLL"]
f=MD["CD94/NKG2A-VLL"]
```

(1) Paired t test for CD94

```python
results1 = DataFrame({'LLL': a, 'VLL': b})
results1.plot(title="CD94",figsize=(10,2))
plt.legend(loc='lower right')
plt.xlabel("Time(ns)")
plt.ylabel("RMSC(Ca)")
plt.show()
p1=stats.ttest_rel(a,b)
print(p1)
p2=stats.mannwhitneyu(a,b)
print(p2)
```

(2) Paired t test for NKG2A

```python
results2 = DataFrame({'LLL': c, 'VLL': d})
results2.plot(title="NKG2A",figsize=(10,2))
plt.legend(loc='lower right')
plt.xlabel("Time(ns)")
plt.ylabel("RMSC(Ca)")
plt.show()
p1=stats.ttest_rel(c,d)
print(p1)
p2=stats.mannwhitneyu(c,d)
print(p2)
```

(3) Paired t test for CD94/NKG2A

```python
results3 = DataFrame({'LLL': e, 'VLL': f})
results3.plot(title="CD94/NKG2A",figsize=(10,2))
plt.legend(loc='lower right')
plt.xlabel("Time(ns)")
plt.ylabel("RMSC(Ca)")
plt.show()
p1=stats.ttest_rel(e,f)
print(p1)
p2=stats.mannwhitneyu(e,f)
print(p2)
```
